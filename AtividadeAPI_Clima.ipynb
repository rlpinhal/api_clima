```python
import requests
import pandas as pd

url = 'https://api.open-meteo.com/v1/forecast'

# parametros utilizados
latitude = float(input('Digite a Latitude desejada: '))
longitude = float(input('Digite a Longitude desejada: '))
quantos_dias = int(input('Digite até quantos dias antes deseja coletar os dados: '))

parametros = {
    'latitude' : latitude,
    'longitude' : longitude,
    'daily' : ['temperature_2m_max','precipitation_sum'],
    'timezone':'auto',
    "past_days" : quantos_dias
}

resposta = requests.get(url, params=parametros)

 #analisar se o link está ativo e puxar os dados do site, definir como json e armazenar
if resposta.status_code == 200:
  dados = resposta.json()

  #criei o dataframe com auxilio da biblioteca Pandas, para exibir com display a tabela
  dados_diarios = dados['daily']
  df_dados_clima = pd.DataFrame(dados_diarios)

  #para adicionar uma coluna nova com a temperatura em Fahrenheit, utiliza-se o seguinte comando:
  df_dados_clima['temp_fahrenheit'] = (df_dados_clima['temperature_2m_max'] * 1.8) + 32

  #com o display, já conseguia resolver, porém quis alterar o formato da data e o nome das colunas. Com o auxilio da IA fiz o seguinte:
  df_dados_clima['time'] = pd.to_datetime(df_dados_clima['time'])
  df_dados_clima.columns = ['Data', 'Temp_Max_C', 'Precipitacao_mm','Temp_Max_F']

  display(df_dados_clima)
else:
  print('Erro ao acessar a API')

#salvar tudo no csv
df_dados_clima.to_csv('relatorio_prof_luiz.csv', index=False)
```
